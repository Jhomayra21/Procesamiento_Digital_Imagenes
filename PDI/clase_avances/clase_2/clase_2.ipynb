```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

#poner imagen 
img= cv2.imread('pato.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#segunda opcion para plotear la imagen
fig,axs = plt.subplots(1,4,figsize=(20,5))

axs[0].imshow(img_rgb)
axs[0].set_title('Imagen Original')

#fig es figura, axs es el arreglo de ejes, imshow es la función para mostrar la imagen, cmap='gray' es para mostrar la imagen en escala de grises
fig_img=axs[1].imshow(img_rgb[...,0],cmap='gray')
#shrink es para reducir el tamaño de la barra de color
fig.colorbar(fig_img, ax=axs[1],shrink=0.5)        
axs[1].set_title('Canal Rojo')      

fig_img=axs[2].imshow(img_rgb[...,1],cmap='gray')
fig.colorbar(fig_img, ax=axs[2],shrink=0.5)        
axs[2].set_title('Canal Verde')      

fig_img=axs[3].imshow(img_rgb[...,2],cmap='gray')
fig.colorbar(fig_img, ax=axs[3],shrink=0.5)        
axs[3].set_title('Canal Azul')      

```

EJERCICIOS PARA GENERAR UNA IMAGEN

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (10, 5)
size_imagen=100
R=np.ones((size_imagen,size_imagen))*255
G=np.ones((size_imagen,size_imagen))*128
B=np.ones((size_imagen,size_imagen))*0

#juntar imagen 
imagen=np.dstack((R,G,B))
imagen=np.array(imagen, dtype=int)

#mostrar imagen
plt.figure()
plt.imshow(imagen)
plt.title('Imagen color : Naranja')
plt.show()
```

ESCALA DE GRISES

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('pato.jpg',cv2.IMREAD_GRAYSCALE)

#valores iniciales de profundidad de color
original=img.dtype
#transformar la imagen a 16 bits
img16bits=np.uint16(img)
img_new=img16bits.dtype
print('profundidad de color :',img_new)

cv2.imwrite('imagen_escala_grises.jpg', img16bits)#guardar con opencv
plt.imsave("imagen_escala_grises_plt.jpg", img, cmap="gray")

#imagen estable tratada
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Imagen Original')
plt.imshow(img, cmap='gray')

plt.subplot(1,2,2)
plt.title('Imagen con 16 bits')
plt.imshow(img16bits, cmap='gray')
```
