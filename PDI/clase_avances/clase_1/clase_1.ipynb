```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [15,9]
img = cv2.imread('pato.jpg', cv2.IMREAD_COLOR)
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cmaps = ['viridis','plasma','inferno','magma','cividis','hot','cool','spring','summer','winter']  # 10 colormaps

fig = plt.figure()
gs = fig.add_gridspec(3,5)

# fila 1: imagen original ocupando las 5 columnas
ax = fig.add_subplot(gs[0,:])
ax.set_title('Imagen Original')
ax.imshow(imgRGB)
ax.axis('off')

# filas 2 y 3: 5 columnas con diferentes cmap aplicados a la imagen en gris
for i, cmap in enumerate(cmaps):
    row = 1 + i//5
    col = i%5
    ax = fig.add_subplot(gs[row, col])
    ax.set_title(cmap)
    ax.imshow(gray, cmap=cmap, vmin=0, vmax=255)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [15,5]
img = cv2.imread('pato.jpg', cv2.IMREAD_COLOR)

imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


r = img[:,:,2]
g = img[:,:,1]
b = img[:,:,0]

#crear imagen por canal 
r_img = np.zeros_like(imgRGB)
r_img [:,:,0] = r

g_img = np.zeros_like(imgRGB)
g_img [:,:,1] = g

b_img = np.zeros_like(imgRGB)
b_img [:,:,2] = b

plt.figure(figsize=(10,5))
plt.subplot(141)
plt.title('Imagen Original')
plt.imshow(imgRGB)

plt.subplot(142) 
plt.title('Canal Rojo')
plt.imshow(r_img)

plt.subplot(143)
plt.title('Canal Verde')
plt.imshow(g_img)

plt.subplot(144)
plt.title('Canal Azul')
plt.imshow(b_img)
```
