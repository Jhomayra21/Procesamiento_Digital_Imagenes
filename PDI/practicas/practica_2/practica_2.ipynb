## PRFUNDIDAD DE COLOR

COLORES EN 4 CUADRANTES

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

size_imagen = 100

imagen = np.ones((size_imagen, size_imagen, 3), dtype=np.uint8) * 255

mitad = size_imagen // 2
imagen[:mitad, :mitad, :] = [255, 128, 0]

imagen[:mitad, mitad:, :] = [255, 0, 0]

imagen[mitad:, :mitad, :] = [0, 255, 0]

imagen[mitad:, mitad:, :] = [0, 0, 255]

# Mostrar imagen
plt.figure()
plt.imshow(imagen)
plt.title('Cuadrantes con np.ones (fondo blanco)')
plt.show()
```

COLORES DE UN CUBO RUBITS (CADA VES QUE EJECUTAMOS CAMBIA DE COLOR)

```python
import matplotlib.pyplot as plt
import numpy as np

size_imagen = 100
rows, cols = 3, 3      
cell_height = size_imagen // rows     
cell_width = size_imagen // cols     

imagen = np.zeros((size_imagen, size_imagen, 3), dtype=np.uint8)

colores = [
    [255, 128,   0],  
    [255,   0,   0],  
    [  0, 255,   0],  
    [  0,   0, 255],  
    [  0,   0,   0],  
    [255, 255,   0],  
    [255,   0, 255],  
    [  0, 255, 255],  
    [128, 128, 128],  
]

index = 0
for i in range(rows):
    for j in range(cols):
        y_start = i * cell_height
        x_start = j * cell_width
        y_end = (i + 1) * cell_height
        x_end = (j + 1) * cell_width if j < cols - 1 else size_imagen 

        imagen[y_start:y_end, x_start:x_end, :] = colores[index]
        index += 1

# Mostrar imagen
plt.figure()
plt.imshow(imagen)
plt.title('Imagen 9 colores (3 filas x 3 columnas)')
plt.axis('off')
plt.show()
```
