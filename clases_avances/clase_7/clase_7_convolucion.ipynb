## CONVOLUCION HORIZONTAL

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def convolucion_media_h(imagen, tkernel):
    kernel = np.ones((1,tkernel),np.float32) / tkernel
    imagen_c=cv2.filter2D(imagen,-1,kernel)
    return imagen_c

img=cv2.imread('imagen3.jpg',0)
img_fin=convolucion_media_h(img,9)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(img_fin,cmap='gray')
plt.title('Convolucion Media Horizontal')
plt.show()
```

## CONVOLUCION HORIZONTAL A COLORES

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def convolucion_media_h(imagen, tkernel):
    kernel = np.ones((1,tkernel),np.float32) / tkernel
    imagen_c=cv2.filter2D(imagen,-1,kernel)
    return imagen_c

img=cv2.imread('imagen3.jpg')
imagen=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_fin=convolucion_media_h(imagen,30)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(imagen,cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(img_fin,cmap='gray')
plt.title('Convolucion Media Horizontal')
plt.show()
```

## CONVOLUCIONES BLUR GAUSS

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

random_img=np.random.randint(0,256,size=(256,256),dtype=np.uint8)
img_gauss=cv2.GaussianBlur(random_img,(9,9),0)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(random_img,cmap='gray')
plt.title('Imagen Aleatoria')
plt.subplot(1,2,2)
plt.imshow(img_gauss,cmap='gray')
plt.title('Imagen con Filtro Gaussiano')
plt.show()
```

## CONVOLUCIONES DE DIFERENTES TIPOS 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Horizontal
def convolucion_horizontal(imagen, k):
    kernel = np.ones((1, k), np.float32) / k
    return cv2.filter2D(imagen, -1, kernel)

# vertical
def convolucion_vertical(imagen, k):
    kernel = np.ones((k, 1), np.float32) / k
    return cv2.filter2D(imagen, -1, kernel)

# diagonal
def convolucion_diagonal(imagen, k):
    kernel = np.zeros((k, k), np.float32)
    np.fill_diagonal(kernel, 1)
    kernel = kernel / k
    return cv2.filter2D(imagen, -1, kernel)

# diagonal inversa 
def convolucion_diagonal_inv(imagen, k):
    kernel = np.zeros((k, k), np.float32)
    kernel = np.fliplr(np.eye(k, dtype=np.float32))
    kernel = kernel / k
    return cv2.filter2D(imagen, -1, kernel)

img = cv2.imread('paisaje.jpg')
imagen = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

k = 30


img_h = convolucion_horizontal(imagen, k)
img_v = convolucion_vertical(imagen, k)
img_d = convolucion_diagonal(imagen, k)
img_di = convolucion_diagonal_inv(imagen, k)

plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
plt.imshow(imagen)
plt.title('Original')
plt.axis('off')

plt.subplot(2,3,2)
plt.imshow(img_h)
plt.title('Horizontal')
plt.axis('off')

plt.subplot(2,3,3)
plt.imshow(img_v)
plt.title('Vertical')
plt.axis('off')

plt.subplot(2,3,4)
plt.imshow(img_d)
plt.title('Diagonal')
plt.axis('off')

plt.subplot(2,3,5)
plt.imshow(img_di)
plt.title('Diagonal inversa ')
plt.axis('off')

plt.tight_layout()
plt.show()
```
