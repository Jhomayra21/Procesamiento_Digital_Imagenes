## CONVOLUCIONES CON MATRIZ KERNEL

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('imagen1.jpg',0)
kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]]) #mascara de convulucion
img_fin=cv2.filter2D(img,-1,kernel) 

plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(img_fin,cmap='gray')
plt.title('Enfoque')
plt.show()

## Nitidez
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('gumball.jpg',0)
kernel = np.ones((5,5),np.float32)/25 
suavizado = cv2.filter2D(img,-1,kernel)
kernel2 = np.ones((7,7),np.float32)/49 
suavizado2 = cv2.filter2D(img,-1,kernel2)

plt.subplot(1,3,1)
plt.imshow(img,cmap='gray')
plt.title('Original')
plt.subplot(1,3,2)
plt.imshow(suavizado,cmap='gray')
plt.title('Suavizado 5x5')
plt.subplot(1,3,3)
plt.imshow(suavizado2,cmap='gray')
plt.title('Suavizado 7x7')
plt.show()

##Suavizado
```

## CONVOLUCION DE PRACTICA (ESCALA DE GRISES)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('imagen1.jpg', 0)

filtros = {
    "Original": None,
    
    "Enfoque": np.array([[0,-1,0],
                         [-1,5,-1],
                         [0,-1,0]]),

    "Desenfoque": np.array([[1,1,1],
                            [1,1,1],
                            [1,1,1]]) / 4,

    "Realce bordes": np.array([[0,0,0],
                               [-1,1,0],
                               [0,0,0]]),

    "Repujado": np.array([[-2,-1,0],
                          [-1,1,1],
                          [0,1,2]]),

    "Deteccion de Bordes": np.array([[0,1,0],
                                     [1,-4,1],
                                     [0,1,0]]),

    "Filtro de tipo Sobel X": np.array([[-1,0,1],
                         [-2,0,2],
                         [-1,0,1]]),

    "Filtro de tipo Sobel Y": np.array([[-1,-2,-1],
                         [0,0,0],
                         [1,2,1]]),

    "Sharpen": np.array([[1,-2,1],
                         [-2,5,-2],
                         [1,-2,1]]),

    "Norte": np.array([[1,1,1],
                       [1,-2,1],
                       [-1,-1,-1]]),

    "Este": np.array([[-1,1,1],
                      [-1,-2,1],
                      [-1,1,1]]),

    "Gauss": np.array([[1,2,3,2,1],
                       [2,7,11,7,2],
                       [3,11,17,11,3],
                       [2,7,11,7,2],
                       [1,2,3,2,1]]) / 115
}

plt.figure(figsize=(15,10))

for i, (nombre, kernel) in enumerate(filtros.items()):
    plt.subplot(4, 4, i+1)
    
    if kernel is None:
        resultado = img
    else:
        resultado = cv2.filter2D(img, -1, kernel)
    
    plt.imshow(resultado, cmap='gray')
    plt.title(nombre)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
