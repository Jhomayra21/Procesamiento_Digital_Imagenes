## NEGATIVO DE UNA IMAGEN

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('imagen1.jpg',cv2.IMREAD_GRAYSCALE)
img_neg=255-img #imagen negativa

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(img_neg,cmap='gray')
plt.title('Negative')
plt.show()
```

## CONTRASTE

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
 
def contraste (img,factor=1.0):
    img=img.astype(np.float32) 
    media=np.mean(img)
    img_contraste=media+(img-media)*factor
    img_contraste=np.clip(img_contraste,0,255).astype(np.uint8)
    return img_contraste    

img=cv2.imread('imagen1.jpg',0) #sin colores quitar el cv2.IMREAD_GRAYSCALE
contrast=7
img_fin=contraste(img,contrast)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(img_fin,cmap='gray')
plt.title('Contraste')
plt.show()
```

## CONTRASTE DE COLORES

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
 
def contraste (img,factor=1.0):
    img=img.astype(np.float32) 
    for i in range(3):
        media=np.mean(img[:,:,i])
        img[:,:,i]=media+(img[:,:,i]-media)*factor
        img[:,:,i]=np.clip(img[:,:,i],0,255)

    return img.astype(np.uint8)

img=cv2.imread('imagen1.jpg') #sin colores quitar el cv2.IMREAD_GRAYSCALE
contrast=9
img_fin=contraste(img,contrast)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) #convertir de BGR a RGB para mostrar correctamente
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img_fin, cv2.COLOR_BGR2RGB), cmap='gray') #convertir de BGR a RGB para mostrar correctamente
plt.title('Contraste')
plt.show()

```

## BRILLO

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def brillo (img,brillo=0):
    img_brillo=img.astype(np.float32)+brillo
    img_brillo=np.clip(img_brillo,0,255).astype(np.uint8)
    return img_brillo


img=cv2.imread('imagen1.jpg',0) 
bril=-70
img_fin=brillo(img,bril)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(img_fin,cmap='gray')
plt.title('Brillo')
plt.show()
```
