## 1. LAPLACIANO

La fórmula del operador Laplaciano es:

$$
\nabla^2 f = \frac{\partial^2 f}{\partial x^2} + \frac{\partial^2 f}{\partial y^2}
$$


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('foto1.jpeg', 0)

laplaciano = cv2.Laplacian(img, cv2.CV_64F)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(laplaciano, cmap='gray')
plt.title('Laplaciano')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('foto1.jpeg', 0)

img = np.float32(img)

d2x = cv2.Sobel(img, cv2.CV_64F, 2, 0)

d2y = cv2.Sobel(img, cv2.CV_64F, 0, 2)

laplaciano = d2x + d2y

plt.figure(figsize=(20,5))

plt.subplot(1,4,1)
plt.imshow(img, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(d2x, cmap='gray')
plt.title('Segunda Derivada X')
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(d2y, cmap='gray')
plt.title('Segunda Derivada Y')
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(laplaciano, cmap='gray')
plt.title('Laplaciano Manual')
plt.axis('off')

plt.show()
```

## 2. LAPLACIANO DEL GAUSSIANO

La fórmula del operador Laplaciano del Gaussiano (LoG) es:

$$
LoG(x, y) = \nabla^2 [G(x, y) * f(x, y)]
$$


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('foto1.jpeg', 0)

gauss = cv2.GaussianBlur(img, (3,3), 0)

log = cv2.Laplacian(gauss, cv2.CV_64F)

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(gauss, cmap='gray')
plt.title('Filtro Gaussiano')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(log, cmap='gray')
plt.title('Laplaciano del Gaussiano')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('foto1.jpeg', 0)

img = np.float32(img)

gauss = cv2.GaussianBlur(img, (5,5), 0)

d2x = cv2.Sobel(gauss, cv2.CV_64F, 2, 0)

d2y = cv2.Sobel(gauss, cv2.CV_64F, 0, 2)

log = d2x + d2y

plt.figure(figsize=(20,5))

plt.subplot(1,4,1)
plt.imshow(img, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(gauss, cmap='gray')
plt.title('Filtro Gaussiano')
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(log, cmap='gray')
plt.title('LoG Manual')
plt.axis('off')

plt.show()
```
