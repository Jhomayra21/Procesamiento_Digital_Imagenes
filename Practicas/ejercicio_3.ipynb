```python
cantidad = int(input("¿Cuántos números vas a ingresar? (mínimo 3): "))

while cantidad < 3:
    cantidad = int(input("Debe ser al menos 3. Ingresa de nuevo: "))

numeros = []

for i in range(cantidad):
    num = float(input(f"Ingrese el número {i+1}: "))
    numeros.append(num)

mayor = max(numeros)
menor = min(numeros)

print("Números ingresados:", numeros)
print("Mayor:", mayor)
print("Menor:", menor)
```
