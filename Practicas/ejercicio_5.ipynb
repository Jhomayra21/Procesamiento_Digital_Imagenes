```python
numeros = []

for i in range(7):
    num = int(input(f"Ingrese el número {i+1}: "))
    numeros.append(num)

frecuencias = {}

for num in numeros:
    if num in frecuencias:
        frecuencias[num] += 1
    else:
        frecuencias[num] = 1

numero_mas_repetido = max(frecuencias, key=frecuencias.get)
cantidad = frecuencias[numero_mas_repetido]

print("Números ingresados:", numeros)
print(f"El número que más se repite es {numero_mas_repetido} y se repite {cantidad} veces")
```
